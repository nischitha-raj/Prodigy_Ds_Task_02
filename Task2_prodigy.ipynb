**CORRELATION HAETMAP**


1.    Converting non-numeric columns to numeric.
2.    Droping non-numeric columns as they are not relevant for correlation.
3.    Calculate correlation matrix and plot heatmap



```python
#Identify non-numeric columns
non_numeric_cols = df.select_dtypes(exclude=['number']).columns

#Convert non-numeric columns to numeric
#Converting 'Gender' column to numeric
if 'Gender' in non_numeric_cols:
    df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})  # Replace with your mapping

#Drop other non-numeric columns as they are not relevant for correlation
df_numeric = df.drop(columns=[col for col in non_numeric_cols if col != 'Gender'])

#Calculate correlation matrix and plot heatmap
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Boxplot**
